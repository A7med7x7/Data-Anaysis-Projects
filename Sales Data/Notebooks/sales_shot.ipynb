```python
customer_data['Order Date'] = pd.to_datetime(customer_data['Order Date'])
average_sales_per_month = customer_data.groupby(customer_data['Order Date'].dt.to_period("M"))['Sales Amount'].mean()
average_sales_moM = average_sales_per_month.pct_change() * 100
plt.figure(figsize=(10, 6))
plt.plot(average_sales_moM.index.to_timestamp(), average_sales_moM.values, label='Month-on-Month Change in Average Sales Amount', marker='o')
plt.xlabel('Order Date')
plt.ylabel('Month-on-Month Change in Sales Amount %')
plt.title('Month-on-Month Change in Average Sales Amount')
plt.legend()

plt.show()
```

# Rolling the 6 month average

```python
import plotly.express as px

customer_data['Order Date'] = pd.to_datetime(customer_data['Order Date'])

rolling_avg_df = customer_data.groupby('CustomerKey')['Sales Amount'].rolling(window=6).mean().reset_index()

fig = px.line(rolling_avg_df, x=customer_data['Order Date'], y=customer_data['Sales Amount'], color='CustomerKey',
              labels={'Sales Amount': 'Rolling 6-Month Avg Sales'}, title='Rolling 6-Month Avg Sales by Customer')
fig.show()
```

# doing the scenario

```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
import matplotlib.pyplot as plt

customer_data['Order Date'] = pd.to_datetime(customer_data['Order Date'])

def calculate_profit(tax_percentage, discount_percentage):
    customer_data['Tax %'] = tax_percentage
    customer_data['Discount %'] = discount_percentage


    customer_data['Profit'] = customer_data['Sales Amount'] * (1 - tax_percentage / 100) * (1 - discount_percentage / 100)
    monthly_profit = customer_data.groupby(customer_data['Order Date'].dt.to_period("M"))['Profit'].sum()


    plt.figure(figsize=(10, 6))
    plt.plot(monthly_profit.index.to_timestamp(), monthly_profit.values, marker='o')
    plt.xlabel('Order Date')
    plt.ylabel('Monthly Profit')
    plt.title('Impact of Tax and Discount on Monthly Profit')
    plt.show()


tax_slider = widgets.FloatSlider(value=5.0, min=0.0, max=10.0, step=0.1, description='Tax %:')
discount_slider = widgets.FloatSlider(value=10.0, min=0.0, max=20.0, step=0.1, description='Discount %:')


interact(calculate_profit, tax_percentage=tax_slider, discount_percentage=discount_slider)
```
